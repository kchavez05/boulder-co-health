# Cleaned Facility Data to Json for Dashboard

```python
#Read the CSV & Create DF
facilities = pd.read_csv("../tableau_facility_data_cities.csv")
facilities.head(10)
```

```python
#transform city names to all caps
facilities["city"] = facilities["city"].str.upper()
```

```python
# add data to name line for display via website code
facilities['facility_name'] = facilities['city']+' - '+facilities['facility_name']
facilities
```

```python
#sort dataframe by facility & city
facilities = facilities.sort_values(by = ['city', 'facility_name'], axis=0, 
                                              ascending = [True,True])
facilities.tail(30)
```

```python
#create facilities DF with data columns
facilities_json = facilities[["city", "facility_name", "facility_id", "rating", "total_ratings", "avg_inspection_score", "avg_violations_count"]].copy()
facilities_json.head()
```

```python
# Export cleaned, dataset for Json conversion 
output_data = "dashboard_facility_data_cities.csv"
facilities_json.to_csv(output_data,header=True,index=False)
```

```python
# function to create json from CSV
def make_json(csvFilePath, jsonFilePath):

    # create the facility dictionary
    data = {"names" : [], "metadata": []}
    
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)

        # Convert each row for the dictionary
        for rows in csvReader:
            facId = rows["facility_id"]
            name = rows["facility_name"]
            data["names"].append(name)
            data["metadata"].append(rows)

    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
#Set the file paths
csvFilePath = r'dashboard_facility_data_cities.csv'
jsonFilePath = r'../Website_Files/web_facility_data.json'
 
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```
