# Cleaned Facility Data to Json for Dashboard

```python
#Read the CSV & Create DF
facilities = pd.read_csv("Export/tableau_facility_data_cities.csv")
facilities.head(10)
```

```python
#transform city names to all caps
facilities["city"] = facilities["city"].str.upper()
```

```python
# add data to name line for display via website code
facilities['facility_name_plus'] = facilities['city']+' - '+facilities['facility_name']+' - '+facilities['siteAddress']
facilities
```

```python
#sort dataframe by facility & city
facilities = facilities.sort_values(by = ['facility_name_plus', 'city'], axis=0, 
                                              ascending = [True, True])
facilities.tail(30)
```

```python
#create facilities DF with data columns
facilities_json = facilities[["city", "facility_name", "facility_name_plus", "facility_id", "rating", "total_ratings", "avg_inspection_score", "avg_violations_count"]]
facilities_json
```

# Merging Database files with Specific Violation data

```python
#Read database export containing Violation information
facilities_violations = pd.read_csv("Export/violations_by_facility.csv")
facilities_violations
```

```python
#sort by descending by most recent violation date
facilities_violations = facilities_violations.sort_values(by = ['inspection_date'], axis=0, 
                                              ascending = False)
facilities_violations
```

```python
#convert Violations cat to Title Case
facilities_violations['violation_category'] = facilities_violations['violation_category'].str.title()
#And lower cat to capital case
facilities_violations['lower'] = facilities_violations['lower'].str.capitalize()
facilities_violations
```

```python
#create the vilations category, contactenate the data, use an underscore to mark the end of one violation and the beginning of another.
facilities_violations['combo'] = facilities_violations['inspection_date']+' - '+facilities_violations['violation_category']+' - '+facilities_violations['lower']+"_"
facilities_violations

# create a simplified DF
facilities_violations_simple = facilities_violations[["facility_id", "combo"]]
facilities_violations_simple
```

```python
# Combine the data into a single dataset
facilities_json_merged = pd.merge(facilities_json, facilities_violations_simple, how="left", on=["facility_id", "facility_id"])

# Flatten the combo column to a single line
merged_vio = facilities_json_merged.groupby('facility_id')['combo'].apply(''.join).reset_index()
merged_vio
```

```python
#copy the df
facilities_json_vio = facilities_json.copy()

#loop over the violations df to assemble FacID plus facility name plus violations
for index, row in merged_vio.iterrows():
        #find the facID and violation data
        facID = merged_vio.loc[index, "facility_id"]
        violations = merged_vio.loc[index, "combo"]

        #drop the violationdata into the clean dataframe
        facilities_json_vio.loc[(facID == facilities_json["facility_id"]), 'violations'] = violations
              
facilities_json_vio_clean = facilities_json_vio[["facility_name_plus", "violations"]]
facilities_json_vio_clean
```

```python
# function to create json from CSV A
def make_json(csvFilePath, jsonFilePath):

    # create the facility dictionary
    data = {"names" : [], "metadata": []}
    
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)

        # Convert each row for the dictionary
        for rows in csvReader:
            name = rows["facility_name_plus"]
            data["names"].append(name)
            data["metadata"].append(rows)

    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
#Set the file paths
csvFilePath = r'Export/dashboard_facility_data_cities.csv'
jsonFilePath = r'../Website_Files/web_facility_data.json'
 
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```

```python
# function to create json from CSV B
def make_json(csvFilePath, jsonFilePath):

    # create the facility dictionary
    data = {"facility_names_plus" : [], "violations": []}
    
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)

        # Convert each row for the dictionary
        for rows in csvReader:
            name = rows["facility_name_plus"]
            data["facility_names_plus"].append(name)
            data["violations"].append(rows)

    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
#Set the file paths
csvFilePath = r'Export/violation_data_dashbaord.csv'
jsonFilePath = r'../docs/web_facility_data2.json'
 
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```
