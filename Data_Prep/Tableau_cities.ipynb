# Imports & Read Data

```python
#Imports
import pandas as pd
import json
import csv
```

```python
#Read the CSV & Create DF
facilities_no_cities = pd.read_csv("tableau_facility_data.csv")
facilities_no_cities.head(10)
```

```python
#Read cleaned data with city information
facilities_cities = pd.read_csv("Data_Prep/Export/unique_restaurants_geocoded.csv")
facilities_cities.head(10)
```

# Matching Facility ID and adding a field to DB data

```python
for index, row in facilities_cities.iterrows():
        #find the city
        facID = facilities_cities.loc[index, "facilityId"]
        city = facilities_cities.loc[index, "city"]

        #drop the city into the clean dataframe
        facilities_no_cities.loc[(facID == facilities_no_cities["facility_id"]), 'city'] = city
              
facilities_no_cities.head()
```

```python
facilities_no_cities.isnull().sum()
```

# Export to CSV

```python
# Export cleaned, final dataset for tableau
output_data = "tableau_facility_data_cities.csv"
facilities_no_cities.to_csv(output_data,header=True,index=False)
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            # Facility Names as the keys
            key = rows["facility_name"]
            data[key] = rows
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
 
# Decide the two file paths according to your
# computer system
csvFilePath = r'dashboard_facility_data_cities.csv'
jsonFilePath = r'web_facility_data_cities.json'
 
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```

```python
# Function to convert a CSV to JSON
# Takes the file paths as arguments
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {"names" : [], "metadata": []}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary
        # and add it to data
        for rows in csvReader:
             
            # Facility Names as the keys
            data["names"].append(rows["facility_name"])
            data["metadata"].append(rows)
 
    # Open a json writer, and use the json.dumps()
    # function to dump data
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
 
# Decide the two file paths according to your
# computer system
csvFilePath = r'dashboard_facility_data_cities.csv'
jsonFilePath = r'web_facility_data.json'
 
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```
