```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

# Import our input dataset
test_df = pd.read_csv('Resources/Restaurant_Inspections_in_Boulder_County.csv')
test_df
```

```python
test_df.dtypes
```

```python
test_df['inspectionDate'] = pd.to_datetime(test_df['inspectionDate'])
test_df.head()
```

```python
filtered_df = test_df[(test_df['inspectionDate']< '2020.01.01')]
filtered_df
```

```python
# Filter out all non-traditional 'restaurants' to include 'CATERING', 'CHURCHES', 'CONCESSIONS SENIOR NUTRITION', 'CONVENIENCE STORES', 'DAY TREATMENT CENTERS', 'GROCERY', 'PREPACKAGED', 'RESIDENTIAL', 'RETAIL COMMESSARY', 'SCHOOLS'
def filter_rows_by_values(filtered_df, col, values):
    return filtered_df[~filtered_df[col].isin(values)]
filtered2_df=filter_rows_by_values(filtered_df, "categoryOfFacility", ["CATERING", "CHURCHES", "CONCESSIONS SENIOR NUTRITION", "CONVENIENCE STORES", "DAY TREATMENT CENTERS", "GROCERY", "PREPACKAGED", "RESIDENTIAL", "RETAIL COMMESSARY", "SCHOOLS"])
filtered2_df
```
